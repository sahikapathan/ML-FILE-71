```python
# cleaning the data set converting totalcharges to numeric

df['totalcharges']= pd.to_numeric(df['totalcharges'], errors='coerce')
```

```python
#droping the missing rows
df = df.dropna()
```

```python
df.shape
```

```python
# encodeing the target variable
df['churn'] = df['churn'].map({'Yes':1,'No':0})
```

```python
# features target sepration
# feature
x=df.drop(['churn','customerid'],axis=1)

# target
y=df['churn']

# here 1removed the unneccessary column customer
#id too
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer



# saclling in processing is not needed for decision tree model
```

```python
cat_cols = x.select_dtypes(include = 'object').columns
num_cols = x.select_dtypes(exclude = 'object').columns
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
# building sklearn encoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat',OneHotEncoder(drop='first',
          handle_unknown='ignore'),cat_cols ),
        ('num','passthrough',num_cols )
    ]
)
```

```python
# now we will create a pipeline end to end for encoder
# one object, no lekage
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
# pipeline = preprocessing + model together
# this aviod the mess up in order

dt_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(random_state=42))
])
```

```python
# train the decision tree


dt_pipeline.fit(x_train,y_train)
```

```python
# 1.Preprocessing (ColumnTransformer): The ColumnTransformer named 'preprocessor' first performs a fit_transform operation on x_train.
# 2.For the categorical columns (cat_cols), the OneHotEncoder learns all unique categories present in x_train. It then transforms these
#  columns into their one-hot encoded representation.
#  3.For the numerical columns (num_cols), the 'passthrough' transformer simply passes them through without any modification.
#  4.The DecisionTreeClassifier learns patterns and relationships from these preprocessed features and the corresponding target variable
#  y_train to build its decision tree model.
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score,classification_report
y_pred_dt = dt_pipeline.predict(x_test)
# test
y_pred_train = dt_pipeline.predict(x_train)
# train
print(accuracy_score(y_train,y_pred_train))

# print(accuracy_score(y_test,y_pred_dt))
# if train acc is high , test acc is low -overfitting
# if train acc is low like(50),test acc is low-undwrfitting
# if train and test acc is good enough and same - properfitting(if difference is 2 then also)
```

```python
print(classification_report(y_test,y_pred_dt))
```

```python
# overfitting intuition
for depth in[2,4,6,10,None]:
  dt_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(
        max_depth=depth,
        random_state=42
        ))
  ])
  dt_pipeline.fit(x_train,y_train)
  train_acc=dt_pipeline.score(x_train,y_train)
  test_acc=dt_pipeline.score(x_test,y_test)
  print(f'Depth:{depth},train={train_acc:.2f},Test={test_acc:.2f}')
```

```python
#feature importance
dt_model=dt_pipeline.named_steps['model']
encoder=dt_pipeline.named_steps['preprocessor']

features_names=encoder.get_feature_names_out()

importances=pd.Series(
    dt_model.feature_importances_,
    index=features_names
).sort_values(ascending=False)

importances.head(10)

# feature importance
 # a pipeline that wraps everything together
 # to inspect internals
 # dt_model - the trained decision tree
 # encoder - the fitted columntransfermer
 # higher value more impoetant
# values sum to 1.0
# trees calculate this automatically during training
```

```python
#random forest

from sklearn.ensemble import RandomForestClassifier

rf_pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('model',RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        max_depth=8,
        n_jobs=-1
    ))
])
rf_pipeline.fit(x_train,y_train)
```

```python
#gradient boosting: Gradient Boosting is another powerful ensemble learning technique that builds
# a model in a stage-wise fashion, similar to Random Forest, but it focuses on correcting the errors of previous models.

from sklearn.ensemble import GradientBoostingClassifier

gb_pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('model',GradientBoostingClassifier(
        n_estimators=150,
        learning_rate=0.1,
        max_depth=6,
        random_state=42
    ))
])
gb_pipeline.fit(x_train,y_train)
gb_pipeline.score(x_test,y_test)
```
