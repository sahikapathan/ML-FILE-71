```python
#train and test split
from sklearn.model_selection import train_test_split
#scikit -learn,it is one of the most popular python libraries for ML
#train and test split is to avoid the data leakage

x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# x_train -input data for training
# x_test -input data for testing
# y_train -correct answers for training
# y_test -correct answers for testing

# x_train + y_train - model learned
# x_test + y_test - model is evaluated
#stratify means spliting training and testing the data
```

```python
x_train = x_train.drop(['loan_id'],axis=1)
x_test = x_test.drop(['loan_id'],axis=1)

#loan id is just an identifier which has no meaning,if we np.kaiser
# if ,the model may learn noise
# removing unnecessary columns from the model train and np.test
```

```python
from re import A
# encoding the model x train and x test
# one hot encode categorical features

from sklearn.preprocessing import OneHotEncoder

categorical_cols=["education","self_employed"]
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_cat=ohe.fit_transform(x_train[categorical_cols])
x_test_cat=ohe.transform(x_test[categorical_cols])

# the drop is to remove the redunant columns to avoid dummy variable trap
# sparse in default would store then as numpy ,
# if false itll give the encoded data as a normal table of numbers instead compressing it
```

```python
# Scaling numeric features:data preprocessing technique used to standardize or normalize the range of independent variables or features in a dataset.
# encoding: refers to the process of converting data from one format to another,converting data into numerical into binary
# StandardScaler: from scikit-learn is a preprocessing class that implements standardization. Standardization (also known as Z-score normalization)
# transforms data such that it has a mean of 0 and a standard deviation of 1.
from sklearn.preprocessing import StandardScaler

numeric_cols=x_train.drop(columns=categorical_cols).columns
scaler=StandardScaler()
x_train_num=scaler.fit_transform(x_train[numeric_cols])
x_test_num=scaler.transform(x_test[numeric_cols])
```

```python
x_train_num_df = pd.DataFrame(x_train_num,columns=numeric_cols,index=x_train.index)
x_train_num_df.head()
x_test_num_df = pd.DataFrame(x_test_num,columns=numeric_cols,index=x_test.index)
x_test_num_df.head()

# x_test_num_df=pd.DataFrame(
#     x_test_num,
#     columns=numeric_cols,
#     index=x_test.index
# )
# x_test_num_df.head()

```

```python
# feature ,combination or concatination
x_train_final=np.hstack([x_train_num_df,x_train_cat])
x_test_final=np.hstack([x_test_num_df,x_test_cat])
```

```python
# # logistic regression is the correct one because the target is categorical output-approved
#  logistic regression is used for classification and predicts problems convert it into class
#  0 and 1,works well as baseline modelit is simple,fast and easy to implement
```

```python
#logistic regression model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train_final,y_train)

# max iter=1000 is for logistic regression to leant using the number
# of iterations(optimization)defaultiteration are not enoough model
# fails to cover this is where the model learns the pattern from the data
```

```python
# evaluate accuracy
from sklearn.metrics import accuracy_score

y_pred=model.predict(x_test_final)
accuracy_score(y_test,y_pred)
```
